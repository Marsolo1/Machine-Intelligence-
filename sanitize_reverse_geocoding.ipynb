```python
import pathlib
import shutil
```

```python
log_file = pathlib.Path("reverse_geocoding.txt")
lines = log_file.read_text().split("\n")[1:-1]
r = list(range(0, len(lines), 3))
for i in r:
    lat, lng = map(float, lines[i].split())
    name = f"{lat:.8f}-{lng:.8f}"
    formatted_addr, label = lines[i + 1], lines[i + 2]
    formatted_addr = formatted_addr[5:]
    label = label[6:]
    print(lat, lng, formatted_addr, label, sep="\n")
    src = pathlib.Path("data/addrs") / label / f"{name}.png"
    if not src.exists():
        print("invalid path:", src.as_posix())
        continue
    if "Hiyoshi" in formatted_addr:
        key = "Hiyoshi"
    elif "Minowacho" in formatted_addr or "Minowachō" in formatted_addr:
        key = "Minowacho"
    elif "Tsunashimahigashi" in formatted_addr:
        key = "Tsunashimahigashi"
    elif "Ida" in formatted_addr:
        key = "Ida"
    else:
        key = f"Others/{formatted_addr}"    
    dest = pathlib.Path(f"data/addrs/{key}/{label}/{name}.png")
    dest.mkdir(parents=True, exist_ok=True)
    shutil.copy(
        src,
        dest
    )
```
